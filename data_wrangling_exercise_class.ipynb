### Step 7: Detect and handle outliers

Pick 2-3 numerical columns (e.g., `trestbps`, `chol`, `thalach`). Create box plots and use the IQR method.

Remember: think about whether outliers are real data or errors before removing them!

```python
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return data[(data[column] < lower) | (data[column] > upper)]
for col in ['RestingBP', 'Cholesterol', 'MaxHR']:
    outliers = detect_outliers_iqr(heart, col)
    print(col, "outliers:", len(outliers))
```

### Step 8: Scale the numerical features

Apply Min-Max scaling to the numerical columns (not the target column).

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

# Identify numeric columns
numeric_cols = heart.select_dtypes(include='number').columns

# Do NOT scale the target column
numeric_cols = numeric_cols.drop('HeartDisease')

# Apply scaling
heart[numeric_cols] = scaler.fit_transform(heart[numeric_cols])

heart.head()

```
